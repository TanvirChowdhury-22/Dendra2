```python
mol5sq = psiresp.Molecule.from_rdkit(
    rdmol,
    charge=0,          
    multiplicity=1,    
    optimize_geometry=True,                      
    conformer_generation_options=dict(           
        n_max_conformers=3,
        keep_original_conformer=True
    ),
)
```

```python
cap_names = ["HN1", "HC3"]  # <-- adjust if needed, e.g. ["HN1", "H32"]

missing = [n for n in cap_names if n not in name2idx]
if missing:
    print("WARNING: not found in PDB:", missing)
cap_indices = [name2idx[n] for n in cap_names if n in name2idx]

constraints = psiresp.ChargeConstraintOptions()
for idx in cap_indices:
    constraints.add_charge_sum_constraint_for_molecule(
        mol5sq, charge=0.0, indices=[idx]
    )

resp_opts = psiresp.resp.RespOptions(exclude_hydrogens=False)
```

```python
job = psiresp.Job(
    molecules=[mol5sq],
    charge_constraints=constraints,    
    resp_options=resp_opts,           
    working_directory="5SQ_resp_no_server"
)
```

```python
job.run()
```
